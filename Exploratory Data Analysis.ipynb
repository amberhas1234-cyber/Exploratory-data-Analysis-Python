### *TASK 2: The cost column in the dataframe “Project” has some missing values. Compute these missing values. Replace the missing values by running average. Use the “For” loop for this task.*

```python
# importing the saved csv files

project_df = pd.read_csv('/kaggle/working/project.csv')
employee_df = pd.read_csv('/kaggle/working/employee.csv')
senior_data_df = pd.read_csv('/kaggle/working/senior.csv')
```

```python
project_df
```

```python
employee_df
```

```python
senior_data_df
```

```python
# calculating running average for missing values

total = 0
for i in range(len(project_df)):
    # using iloc to access values of 'Cost' column
    if pd.isna(project_df.iloc[i, 2]):
        average = total/(i)
        project_df.iloc[i, 2] = average
        total += project_df.iloc[i, 2]
    else:
        total += project_df.iloc[i, 2]
```

### *Task 3: Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.*

```python
# creating new columns 

employee_df['First Name'] = employee_df['Name'].str.split(' ').str[0]
employee_df['Last Name'] = employee_df['Name'].str.split(' ').str[1]
```

```python
# dropping old 'Name' column

employee_df.drop(columns=['Name'], inplace=True)
```

```python
employee_df
```

### *Task 4: Join all three dataframes in one single dataframe. Name it “Final”*

```python
# joining the dataframes

df = pd.merge(project_df, employee_df, on='ID', how='inner')
final = pd.merge(df, senior_data_df, on='ID', how='inner')
```

```python
final
```

### *Task 5: A new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.*

```python
# using numpy for creating new column based on condition
final['Bonus'] = np.where(final['Status'] == 'Finished', final['Cost']*0.05, np.nan)
```

```python
final
```

### *Task 6: Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.*

```python
# demoting designation level by 1 i.e. Designation Level + 1

final['Designation Level'] = np.where(
    final['Status'] == 'Failed', 
    final['Designation Level'] + 1, 
    final['Designation Level']
)
```

```python
final
```

```python
# deleting employee records for designation level > 4

records_to_delete = final[final['Designation Level'] > 4].index
final = final.drop(records_to_delete)
```

```python
final
```

### *Task 7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.*

```python
# Adding 'mr' and 'mrs' to the first name

final['First Name'] = np.where(
    final['Gender'] == 'M', 
    'Mr. ' + final['First Name'], 
    'Mrs. ' + final['First Name']
)

final
```

```python
# dropping gender column

final.drop(columns=['Gender'], inplace=True)
```

```python
final
```

### *Task 8: Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.*

```python
# Promote designation level by 1 if age > 29 i.e. Desgination Level - 1

final['Designation Level'] = final.apply(
    lambda x: x['Designation Level']-1 if x['Age'] > 29 else x['Designation Level'],
    axis=1
)

final
```

### *Task 9: Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost*

```python
# Total Cost of projects per employee

TotalProjCost = final.groupby(['ID', 'First Name']).agg({'Cost': 'sum'}).reset_index()
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)
TotalProjCost
```
